# Testing The Model using Evaluation Metrics for Regression Algorithms

```python
from sklearn import metrics
```

Mean Square Error-Calculates the Distance between actual y value and predicted y values.
MSE is less means The model is Good.

```python
print(metrics.mean_squared_error(y_test,y_pred))
```

Root Mean Squared Error is the square root of Mean Squared Error.

```python
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

R2 score 
When this score is high then the model is good

```python
print(metrics.r2_score(y_test,y_pred))
```

```python
MLR_train=MLR.predict(x_train)
```

```python
print(metrics.r2_score(y_train,MLR_train))
```

# In Linear Regression model the testing and training accuracy are nearly equal,so it gives a TradeOff solution.It has Low accuracy so we will try with another Regression Algorithm.
